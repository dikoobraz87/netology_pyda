Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
medv = np.array(data['MEDV']).reshape(-1,1)
lstat = np.array(data['LSTAT']).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
# x = lstat,
# y = medv
X_train, X_test, y_train, y_test = train_test_split(
    lstat,
    medv,
    test_size = 0.2
)
```

```python
model.fit(X_train, y_train)
```

```python
predicted_y = model.predict(X_test)
```

```python
model.score(X_test, y_test)
```

```python
plt.xlabel('LSTAT')
plt.ylabel('MEDV')
plt.title('Линейная регрессия')
x = np.linspace(0, 55, 100)
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(0,40)
plt.ylim(0,60)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
fig=plt.gcf()
fig.set_size_inches(11,11)
plt.show()
```

# Дискретное распределение случайной величины

Напишите код для моделирования 10 подбрасываний игрального кубика. Начертите диаграму histogram. Какое это распределение?

```python
from random import randint
```

```python
number_of_randoms = 10
plt.hist(np.random.randint(1, 7, number_of_randoms), 10)
plt.show()
```

**такое распределение близко к равномерному**

моделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
number_of_randoms_1 = 100
number_of_sums = 1000
X1 = np.zeros([number_of_sums])
for i in range(number_of_sums):
    X1[i] = np.sum(np.random.randint(1, 7, number_of_randoms_1))
plt.hist(X1, 50)
plt.xlabel('X values')
plt.ylabel('Number of elements in each X')
plt.show()
```

**график принял вид стремящегося к нормальному распределению по логике центральной предельной теоремы, так как мы сумировали большое количество выборок с разным распределением**

### Другой вариант решения 
...более классическими методами

```python
toss = [randint(1, 6) for i in range(10)]
toss = pd.DataFrame(toss)
toss.hist()
```

```python
def toss_cube(repeat_cycle, repeat_number):
    toss_list = []
    for i in range(repeat_cycle):
        i = sum([randint(1, 7) for x in range(repeat_number)])
        toss_list.append(i)
    toss_df = pd.DataFrame(toss_list)
    return toss_df.hist()
```

```python
toss_cube(1000, 100)
```
