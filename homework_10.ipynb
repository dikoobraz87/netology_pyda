### Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
rating_data = pd.read_csv('ratings.csv')
movie_data = pd.read_csv('movies.csv')
```

```python
rating_data = rating_data[ rating_data.rating == 5.0 ]
```

```python
rating_data = rating_data.groupby('movieId').count().sort_values('rating', ascending=False)
```

```python
joined_data = rating_data.merge(movie_data, on='movieId', how='left')
joined_data = joined_data[['movieId', 'title']]
```

```python
joined_data.head(1)
```

```python
movies = joined_data['title']
top_movies = movies[0]
print(f'{top_movies} - получил наибольшее количество оценок 5.0')
```

### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data_power = pd.read_csv('power.csv')
```

```python
data_power.head()
```

```python
data_power[ (data_power['country'].str.contains('lat', case=False)) | 
           (data_power['country'].str.contains('est', case=False)) | 
           (data_power['country'].str.contains('lit', case=False)) 
          ]['country'].unique()
```

```python
data_power = data_power[
    ((data_power['category'] == 4) | (data_power['category'] == 12) | (data_power['category'] == 21)) & 
    ((data_power['country'] == 'Latvia') | (data_power['country'] == 'Lithuania') | (data_power['country'] == 'Estonia')) & 
    (data_power['quantity'] > 0) & 
    (data_power['year'] >= 2005) & (data_power['year'] <= 2010)   
]
data_power.head()
```

```python
print('Суммарное потребление =', sum(data_power['quantity']))
```
